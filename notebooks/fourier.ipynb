```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python

w =  5

x = np.linspace(0, np.pi*2, 500)
y = lambda t: signal.sawtooth(w * t)

plt.plot(x, y(x))
#plt.plot(x, np.sin(w * x))


reihe = -0.636619772367582*np.sin(w*x) -0.318309886183791*np.sin(w*2*x) -0.2122065907891942*np.sin(w*3*x) -0.1591549430918955*np.sin(w*4*x) -0.12732395447351677*np.sin(w*5*x)
plt.plot(x, reihe)
```
